# Simplified Tracer in R

author: Walter Xie

date: 20 May 2016

This package is a simplified implementation of 
[Tracer v1.6](http://beast.bio.ed.ac.uk/tracer) in R.
But it is not only the subset, 
but also aims to provide more visualizations to understand the Bayesian MCMC result, 
and also to make batch processing of multipe logs easier. 

**Citation:** Rambaut A, Suchard MA, Xie D & Drummond AJ (2014) Tracer v1.6

**Dependendcy:** [ComMA](https://github.com/walterxie/ComMA) package

```r
suppressMessages(library("TraceR"))
```

## Data
A \*BEAST 2 log is used as the example. 

```r
mcmc.log <- readMCMCLog("data/star.beast.log")
```

```r
names(mcmc.log)
```

```r
burn.in.stats <- getBurnIn(rownames(mcmc.log))
```

## Plot density

## Plot traces

```r
gt <- gtTraces(mcmc.log, burn.in.stats$burn.in, verbose = F)
plot(gt)
```

```r
gt <- gtTraces(mcmc.log, burn.in.stats$burn.in, 
               trace.id=c("TreeHeight.Species", "TreeHeight.t:tree_0_0", 
                          "TreeHeight.t:tree_0_1"), 
              verbose = F)
plot(gt)
```
